## Functions

```python
def confusion_mat(predictions, y_test, threshold = 0.5):
    predictions_class = [0  if i > threshold else 1 for i in predictions]
    mat = confusion_matrix(y_test, predictions_class)

    sns.heatmap(mat, square=True, annot=True, cbar=False)
    plt.xlabel('predicted value')
    plt.ylabel('true value')
    return mat
```

```python
def precision_recall(mat):
    true_pos = mat[1][1]
    true_neg = mat[0][0]
    false_pos = mat[0][1]
    false_neg = mat[1][0]

    precision = true_pos / (true_pos+false_pos)
    recall = true_pos / (true_pos + false_neg)
    
    print('precision: %0.2f, recall: %0.3f' %(precision, recall))
    return 0
```

```python
def pr_curve(predictions, y_test):
    average_precision = average_precision_score(y_test, predictions)

    print('Average precision-recall score: {0:0.3f}'.format(
          average_precision))

    a4_dims = (11.7, 8.27)
    fig, ax = plt.subplots(figsize=a4_dims)
    precision, recall, _ = precision_recall_curve(y_test, predictions)

    step_kwargs = ({'step': 'post'}
                   if 'step' in signature(plt.fill_between).parameters
                   else {})
    plt.step(recall, precision, color='b', alpha=0.2,
             where='post')
    plt.fill_between(recall, precision, alpha=0.2, color='b', **step_kwargs)

    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title('2-class Precision-Recall curve: AP={0:0.3f}'.format(
              average_precision))
```

```python
def roc(predictions, y_test):
    fpr, tpr, thresholds = roc_curve(y_test, predictions)
    roc_auc = auc(fpr, tpr)
    lw = 2
    plt.figure()
    a4_dims = (11.7, 8.27)
    fig, ax = plt.subplots(figsize=a4_dims)

    plt.plot(fpr, tpr, color='darkorange', lw=1, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic')
    plt.legend(loc="lower right")
    plt.show()
    return 0
```

```python
def plotLiftChart(actual, predicted):
    df_dict = {'actual': list (actual), 'pred': list(predicted)}
    df = pd.DataFrame(df_dict)
    pred_ranks = pd.qcut(df['pred'].rank(method='first'), 100, labels=False)
    actual_ranks = pd.qcut(df['actual'].rank(method='first'), 100, labels=False)
    pred_percentiles = df.groupby(pred_ranks).mean()
    actual_percentiles = df.groupby(actual_ranks).mean()
    plt.title('Lift Chart')
    plt.plot(np.arange(.01, 1.01, .01), np.array(pred_percentiles['pred']),
             color='darkorange', lw=2, label='Prediction')
    plt.plot(np.arange(.01, 1.01, .01), np.array(pred_percentiles['actual']),
             color='navy', lw=2, linestyle='--', label='Actual')
    plt.ylabel('Target Percentile')
    plt.xlabel('Population Percentile')
    plt.xlim([0.0, 1.0])
    plt.ylim([-0.05, 1.05])
    plt.legend(loc="best")
```

```python
def data_prep(data):
    data0 = data[data.AMI_FLAG==0]
    data1 = data[data.AMI_FLAG==1]
    
    n0_test, n1_test = 19800,550 
    data0_test = data0.sample(n=n0_test)
    data1_test = data1.sample(n=n1_test)
    
    #-----
    data0 = data0.drop(data0_test.index)
    data1 = data1.drop(data1_test.index)
    
    test_frames = [data0_test,data1_test]
    test = pd.concat(test_frames)
    test.fillna(-1,inplace=True)
    
    
    print('Testing/Holdout Set:',len(test),'0:',len(data0_test),'1:',len(data1_test))
    #-----
    
    n0_vald, n1_vald = 19800,550
    data0_vald = data0.sample(n=n0_vald)
    data1_vald = data1.sample(n=n1_vald)
    
    #-----
    data0 = data0.drop(data0_vald.index)
    data1 = data1.drop(data1_vald.index)
    
    valid_frames = [data0_vald,data1_vald]
    valid = pd.concat(valid_frames)
    valid.fillna(-1,inplace=True)
    
    print('Validation Set:',len(valid),'0:',len(data0_vald),'1:',len(data1_vald))
    #-----
    
    batches = 35
    
    step = int(len(data0)/batches)
    
    train1 = data1
    
    train = list()
    
    for i in range(batches):
        if i != batches-1:
            temp = data0[i*step:(i+1)*step]
            train_frames = [temp,train1]
            temp2 = pd.concat(train_frames)
            train.append(temp2)
            print('Training Set',i+1,':',len(temp2),'0:',len(temp),'1:',len(train1))
        else:
            temp = data0[i*step:]
            train_frames = [temp,train1]
            temp2 = pd.concat(train_frames)
            train.append(temp2)
            print('Training Set',i+1,':',len(temp2),'0:',len(temp),'1:',len(train1))
    
    return(train, valid, test)
```

```python
train, valid, test = data_prep(data)
```

```python
def process_data(data):
    data.fillna(-1,inplace=True)
    cols = data.columns
    
    X_train = data.drop([cols[0],cols[3]],axis=1)
    y_train = data[cols[3]]
    X_valid = valid.drop([cols[0],cols[3]],axis=1)
    y_valid = valid[cols[3]]
    
    return(X_train,y_train,X_valid,y_valid)
```

```python
models = list()

preds_valid = list()
preds_valid_proba = list()

for i in range(len(train)):
    print('### Training for Dataset:',i+1)
    X_train,y_train,X_valid,y_valid = process_data(train[i])
    cols = list(X_train.columns)
    
    # check for categorical columns 
    cols_cat = list(set(cols) - set(X_train._get_numeric_data().columns))
    col_cat_indices = [cols.index(cols_cat[i]) for i in range(len(cols_cat))]
    
    model = CBC(iterations=300, depth=6, learning_rate=0.06, loss_function= 'Logloss')
    model.fit(X_train, y_train, cat_features = col_cat_indices, plot=False, early_stopping_rounds = 50, silent=True);
    models.append(model)
    
    pred_train = model.predict(X_train)
    pred_train_proba = model.predict_proba(X_train)
    print('Accuracy: Training', accuracy_score(y_train, pred_train), )
    
    pred_valid = model.predict(X_valid)
    pred_valid_proba = model.predict_proba(X_valid)
    print('Accuracy: Validation Set',accuracy_score(y_valid, pred_valid))
    print('Precision: Validation Set',precision_score(y_valid, pred_valid))
    print('Recall: Validation Set',recall_score(y_valid, pred_valid))
    print('F1 Score: Validation Set',f1_score(y_valid, pred_valid))
    
    preds_valid.append(pred_valid)
    preds_valid_proba.append(pred_valid_proba)
```

```python
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    #print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```
